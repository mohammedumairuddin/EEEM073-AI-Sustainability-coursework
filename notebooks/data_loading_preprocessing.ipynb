```python
import pandas as pd

# Used this to directly use dataset from the repo
url = "https://raw.githubusercontent.com/mohammedumairuddin/EEEM073-AI-Sustainability-coursework/refs/heads/main/datasets/crop_production.csv"
df = pd.read_csv(url)
df.head()
```

```python
df["Yield"] = df["Production"] / df["Area"]
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = ["State_Name","District_Name","Season","Crop"]

for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
import os

# Create the directory if it does not exist
os.makedirs('data', exist_ok=True)
df.to_csv("data/processed_dataset.csv", index=False)
```
